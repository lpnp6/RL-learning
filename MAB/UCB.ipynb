上置信界算法(Upper Confidence Bound, UCB)是一种基于不确定性的算法，理论来自于霍夫丁不等式：$$\mathbb{P} \left\{ \mathbb{E}[X] \geq \bar{x}_n + u \right\} \leq e^{-2nu^2} $$
$X_{1}, X_{2}, X_{3} \dots ,X_{n}$ 为n个独立同分布的随机变量，经验期望为 $x_{n}=\frac{1}{n}\sum{X_{i}}$

将$\hat{Q}_{t}(a)$代入$\bar{x}_n$，$\hat{U}_{t}(a)$代入$u$，令$p=e^{{-2N_{t}\hat{U}^{2}_{t}(a)}}$，所以$Q_{t}(a) < \hat{Q}_{t}(a) + \hat{U}_{t}(a)$的概率为$1-p$。当$p$很小时，$Q_{t}(a) < \hat{Q}_{t}(a) + \hat{U}_{t}(a)$的概率就非常大，$\hat{Q}_{t}(a) + \hat{U}_{t}(a)$就为期望奖励上界。选取期望奖励上界最大的杆，即$a = \operatorname*{arg\,max}_{a \in \mathcal{A}} (\hat{Q}(a) + \hat{U}(a))$。

令$p=\frac{1}{t}$，则令$\hat{U}(a)=\sqrt{{\frac{\log t}{2(N_t(a)+1)}}}$。
