$\epsilon$ 贪婪算法每次以概率 $1-\epsilon$ 选择以往经验中期望奖励估值最大的那根拉杆，以概率 $\epsilon$ 随机选择一根拉杆

$
 a_t = 
\begin{cases} 
\arg\max_{a \in \mathcal{A}} \hat{Q}(a), & \text{采样概率：} 1-\epsilon \\
\text{从 } \mathcal{A} \text{ 中随机选择}, & \text{采样概率：} \epsilon 
\end{cases} $

```python
from utils import plot_results
from EpsilonGreedy import EpsilonGreedy
from MAB import BernoulliBandit
import numpy as np
```

```python
np.random.seed(0)
epsilon_greedy_solver = EpsilonGreedy(BernoulliBandit(10), epsilon=0.01)
epsilon_greedy_solver.run(100000)
print('epsilon-贪婪算法的积累懊悔为：', epsilon_greedy_solver.regret)
plot_results([epsilon_greedy_solver], ["EpsilonGreedy"])
```

```python
np.random.seed(0)
epsilons = [1e-4, 0.01, 0.1, 0.25, 0.5]
epsilon_greedy_solver_list = [EpsilonGreedy(BernoulliBandit(10), epsilon=e)for e in epsilons]
epsilon_greedy_solver_names = ["epsilon={}".format(e) for e in epsilons]
for solver in epsilon_greedy_solver_list:
    solver.run(100000)

plot_results(solvers=epsilon_greedy_solver_list, solver_names=epsilon_greedy_solver_names)
```

无论 $\epsilon$ 取值多少，累积懊悔都是线性增长的。
